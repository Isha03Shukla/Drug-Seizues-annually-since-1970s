```python
sns.set_style('whitegrid')
plt.figure(figsize=(35,15))
sns.countplot(x="Drug Group", data=data, palette='cubehelix');
```

```python
mean_col = ['Region','Country', 'Drug Group', 'Drug',
       'Year']
sns.set(rc = {'figure.figsize':(15,8)})
sns.pairplot(data[mean_col],hue = 'Drug Group', palette='Blues')
```

```python
mean_col = ['Region','Country', 'Drug Group', 'Drug',
       'Year']
sns.set(rc = {'figure.figsize':(10,10)})
sns.pairplot(data[mean_col],hue = 'Region', palette='YlOrBr')
```

```python
plt.figure(figsize=(100,150))
plt.hist(data['Year'])
plt.title("Drugs")
plt.xlabel("Year")
plt.ylabel("Drug Group")
plt.show()
```
